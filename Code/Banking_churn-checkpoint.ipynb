```python
# 1. IMPORT 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Load Data
df = pd.read_csv("C:/Users/rahul/Downloads/banking_churn/Bank_Churn.csv")
```

```python
#basic checks 
print("Rows :", df.shape[0])
```

```python
print("Columns :", df.shape[1])
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
# Overall Churn Rate 

total_customers = len(df)
churned = df['Exited'].sum()
reatined = total_customers - churned
churn_rate = round(churned/total_customers * 100, 2)

print("Total Customers :", total_customers )
print("Churned :", churned )
print("Retained Customers :", reatined )
print("Churn Rate :", churn_rate, "%" )

sns.countplot(x ='Exited', data=df, hue='Exited', palette= ['green','red'])
plt.title('Churn Distribution')
plt.show()
```

```python
# churn by geography

geo = df.groupby('Geography')['Exited'].mean()*100
print(geo.round(2))

sns.barplot(x=geo.index, y=geo.values,hue = geo.index, palette= 'Reds')
plt.title('Churn % by Geography')
plt.xlabel('Geography')
plt.ylabel('Churn %')
```

```python
# churn by age group

df['Age_Group'] = pd.cut(df['Age'], bins= [0,30,40,50,60,100], labels=['<30','31 - 40','41 - 50','51 - 60','60+'])
age =df.groupby('Age_Group',observed= True)['Exited'].mean()*100

print(age.round(2))

sns.barplot(x= age.index, y= age.values, hue=age.index, palette= 'Oranges_d')
plt.title('Churn Rate by Age Group')
plt.ylabel('Churn Rate %')
```

```python
# churn rate - active vs inactive

active = df.groupby('IsActiveMember')['Exited'].mean()*100
active.index = ['Inactive','Active']
print(active.round(2))

sns.barplot(x=active.index, y=active.values, hue= active.index, palette = 'Blues_d')
plt.title('Churn rate - Active vs Inactive')
plt.ylabel('Churn Rate %')
```

```python
# churn rate by Number of Products 

prod = df.groupby('NumOfProducts')['Exited'].mean() * 100
print(prod.round(2))

sns.barplot(x=prod.index, y=prod.values, hue = prod.index, palette='Purples_d')
plt.title('Churn Rate by Number of Products')
plt.ylabel('Churn Rate %')
```

```python
#  balance for churned vs retained

bal = df.groupby('Exited')['Balance'].mean().round(2)
bal.index = ['Retained', 'Churned']
print(bal)

sns.boxplot(x='Exited', y='Balance', data=df, hue= 'Exited', palette=['green','red'])
plt.title('Balance - Churned vs Retained')
```

```python
# Churn rate per gender
gen = df.groupby('Gender')['Exited'].mean() * 100
print(gen.round(2))

sns.barplot(x=gen.index, y=gen.values, hue = gen.index, palette='Set2')
plt.title('Churn Rate by Gender')
plt.ylabel('Churn Rate %')
```

```python
# credit score  to churn 
cr = df.groupby('Exited')['CreditScore'].mean().round(2)
cr.index = ['Retained', 'Churned']
print(cr)

sns.boxplot(x='Exited', y='CreditScore', data=df, hue= 'Exited', palette=['green','red'])
plt.title('Credit Score - Churned vs Retained')
```

```python
# GERMANY HIGH RISK SEGMENT
# Who is the single highest risk customer?
```

```python
#Type 1
# Germany + inactive + (41-50 or 51-60 ) + 3-4 products
risk_type1 = df[
    (df['IsActiveMember'] == 0) &
    (df['Geography'] == 'Germany') &
    (df['Age_Group'].isin(['41 - 50', '51 - 60'])) &
    (df['NumOfProducts'].isin([3, 4]))]

rate1 = round(risk_type1['Exited'].mean() * 100, 2)
print("\nOver-sold Segment")
print("Total :", len(risk_type1))
print("Churned :", risk_type1['Exited'].sum())
print("Churn Rat :", rate1, "%")
print("Avg Balance :", risk_type1['Balance'].mean())
```

```python
#Type 2 
## Germany + inactive + (41-50 or 51-60 ) + 1 products
risk_type2 = df[
    (df['IsActiveMember'] == 0) &
    (df['Geography'] == 'Germany') &
    (df['Age_Group'].isin(['41 - 50', '51 - 60'])) &
    (df['NumOfProducts'] == 1)
]

rate2 = round(risk_type2['Exited'].mean() * 100, 2)
print("\nno loyalty segment")
print("total :", len(risk_type2))
print("churned :", risk_type2['Exited'].sum())
print("churn rate :", rate2, "%")
print("avg balance :", risk_type2['Balance'].mean())
```

```python
#FUNNEL 2 — AGE GROUP 
#Question  -- Is this an age problem everywhere -- or is it concentrated somewhere?
# Goal  : Find the specific high risk profile

age = df.groupby('Age_Group', observed=True)['Exited'].mean() * 100
print("global churn by age group")
print(age.round(2))

sns.barplot(x=age.index, y=age.values, hue= age.index, palette='Oranges_d')
plt.title('Global Churn by Age Group')
plt.ylabel('Churn Rate %')
```

```python
# How bad is each age group globally?
age_geo = df.groupby(['Age_Group', 'Geography'], observed=True)['Exited'].mean() * 100
print("\nage group churn by country")
print(age_geo.round(2))

age_geo_df = age_geo.reset_index()
age_geo_df.columns = ['Age_Group', 'Geography', 'Churn_Rate']

sns.barplot(x='Age_Group', y='Churn_Rate', hue='Geography', data=age_geo_df, palette = 'Blues_d' )
plt.title('Age Group Churn by Country')
plt.ylabel('Churn Rate %')
```

```python
# Do older customers with more products churn more?
age_prod = df.groupby(['Age_Group', 'NumOfProducts'], observed=True)['Exited'].mean() * 100
print("\nage group churn by products")
print(age_prod.round(2))

age_prod_df = age_prod.reset_index()
age_prod_df.columns = ['Age_Group', 'NumOfProducts', 'Churn_Rate']

sns.barplot(x='Age_Group', y='Churn_Rate', hue='NumOfProducts', data=age_prod_df, palette = 'Blues')
plt.title('Age Group Churn by Products')
plt.ylabel('Churn Rate %')
```

```python
# Are inactive older customers the core problem?
age_active = df.groupby(['Age_Group', 'IsActiveMember'], observed=True)['Exited'].mean() * 100
print("\nage group churn by activity")
print(age_active.round(2))

age_active_df = age_active.reset_index()
age_active_df.columns = ['Age_Group', 'IsActiveMember', 'Churn_Rate']
age_active_df['IsActiveMember'] = age_active_df['IsActiveMember'].map({0: 'Inactive', 1: 'Active'})

sns.barplot(x='Age_Group', y='Churn_Rate', hue='IsActiveMember', data=age_active_df, palette = 'BuGn')
plt.title('Age Group Churn by Activity')
plt.ylabel('Churn Rate %')
```

```python
# COMBINE ALL
# Final high risk age profile

age_risk = df[
    (df['IsActiveMember'] == 0) &
    (df['Age_Group'].isin(['41 - 50', '51 - 60'])) &
    (df['NumOfProducts'].isin([3, 4]))]

rate = round(age_risk['Exited'].mean() * 100, 2)
print("\nage funnel — high risk profile")
print("segment  : inactive + 41-60 + 3-4 products")
print("total    :", len(age_risk))
print("churned  :", age_risk['Exited'].sum())
print("churn rate :", rate, "%")
print("avg balance :", age_risk['Balance'].mean().round(2))
print("overall churn :", 20.37, "%")
print("risk multiplier :", round(rate / 20.37, 1), "x higher")
```
